```python
X = df.drop('Wine',axis=1)
y = df['Wine']
```

```python
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y = train_test_split(X,y,stratify=y,test_size=0.2,random_state=123456789)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver='newton-cg',max_iter=120,verbose=0)
```

```python
model = lr.fit(train_x,train_y)
```

```python
y_pred = model.predict(test_x)
```

```python
from sklearn import metrics

metrics.accuracy_score(test_y,y_pred)
```

```python
cm = metrics.confusion_matrix(test_y,y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=['1','2','3'])
cm_display.plot(cmap='Blues_r')
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(lr,train_x,train_y).mean()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer,KBinsDiscretizer,MinMaxScaler,StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
import numpy as np

pipe_clf = Pipeline([
  (
    "FeatureEngineering",
    ColumnTransformer(
        [
          ("alcohol", "drop", ["Alcohol"]),
          ("malic_acid", MinMaxScaler(), ["Malic.acid"]),
          ("magnesium", 
            Pipeline([
                      ("LogTransformation", FunctionTransformer(lambda value: np.log1p(value)) ),
                      ("Kbins", KBinsDiscretizer(n_bins=4) )
            ]),
            ["Mg"]
          ),
          ("nonflavanoid_phenols", "passthrough", ["Nonflavanoid.phenols"]),
        ],
      remainder=StandardScaler()
    )
  ),
  (
    "PCA_red",
    PCA(n_components=10)  
  ),
  (
    "RNDF_clf",
    RandomForestClassifier(n_estimators=250, max_depth=40, random_state=214)
  )
])
```

```python
pipe_clf
```

```python
model = pipe_clf.fit(train_x,train_y)
```

```python
predicted = model.predict(test_x)
```

```python
metrics.accuracy_score(test_y,predicted)
```

```python
cross_val_score(pipe_clf,train_x,train_y,cv=5).mean()
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

ohe = OneHotEncoder(sparse_output=False)
train_y_ohe = ohe.fit_transform(np.array(train_y).reshape(-1,1))
test_y_ohe = ohe.transform(np.array(test_y).reshape(-1,1))
```

```python
model = Sequential([
    keras.layers.Dense(10,activation='relu',input_shape=(13,)),
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(3,activation='softmax')
])
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(x=train_x,y=train_y_ohe,validation_split=0.2,epochs=50)
```
